```python
import matplotlib.pyplot as plt
import numpy as np
```

Q1

```python
def make_sine_wave(x, A, w):
    return A * np.sin(w * x)
```

```python
x = np.linspace(0, 2 * np.pi, 1000)
amplitudes = [0.5, 1, 1.5, 2, 2.5]
frequencies = [1, 2, 3, 4, 5]
line_styles = ['-', '--', '-.', ':', (0, (3, 1, 1, 1))]
```

```python
plt.figure(figsize=(10,6))
for A, w, style in zip(amplitudes, frequencies, line_styles):
    y = make_sine_wave(x, A, w)
    plt.plot(x, y, linestyle=style, label=f"A={A}, w={w}")
plt.title("Different Sine Waves")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
df = pd.read_csv("stars.csv")
```

```python
print("First 5 rows:")
print(df.head())

print("\nShape (rows, columns):")
print(df.shape)

print("\nColumn names and data types:")
print(df.dtypes)
```

```python
average_mass = df['Mass (M)'].mean()
average_temp = df['Temperature (K)'].mean()
print(f"\nAverage mass: {average_mass:.2f}")
print(f"Average temperature: {average_temp:.2f}")

largest_radius_star = df.loc[df['Radius (R)'].idxmax()]
print("\nStar with the largest radius:")
print(largest_radius_star)

m_type_stars = df[df['Spectral_Type'].str.startswith('M')]
print(f"\nNumber of M-type stars: {len(m_type_stars)}")
```

```python
closest_stars = df.sort_values(by='Distance (ly)').head(3)
print("\n3 Closest Stars:")
print(closest_stars)
```

```python
m_type_stars.to_csv("m_type_stars.csv", index=False)
print("\nFiltered M-type stars saved to 'm_type_stars.csv'")
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df['Temperature (K)'], bins=20)
plt.title("Temperature Distribution of Stars")
plt.xlabel("Temperature (K)")
plt.ylabel("Number of Stars")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
penguins = sns.load_dataset("penguins")

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

sns.scatterplot(data=penguins, x="bill_length_mm", y="flipper_length_mm",
                hue="species", style="species", ax=ax1)
ax1.set_title("Flipper Length vs. Bill Length")
ax1.set_xlabel("Bill Length (mm)")
ax1.set_ylabel("Flipper Length (mm)")
ax1.legend(title="Species")
ax1.grid(True)

sns.histplot(data=penguins, x="body_mass_g", kde=True, bins=20, ax=ax2)
ax2.set_title("Distribution of Penguin Body Mass")
ax2.set_xlabel("Body Mass (g)")
ax2.set_ylabel("Count")
ax2.grid(True)

plt.show()
```
