```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.optimize as fit
import scipy.constants as cs
```

## Data Processing and Reduction

```python
df_min = pd.read_csv("dataset_min.txt", sep=" ", header=None)
df_min.iloc[0:1830, 2] = df_min.iloc[0:1830, 3]
df_min.iloc[0:1830, 3] = df_min.iloc[0:1830, 4]
df_min.iloc[0:1830, 4] = df_min.iloc[0:1830, 5]
df_min.iloc[0:1830, 5] = df_min.iloc[0:1830, 6]
df_min.iloc[0:1830, 6] = df_min.iloc[0:1830, 7]
df_min.iloc[0:1830, 7] = df_min.iloc[0:1830, 8]
df_min.iloc[0:1830, 8] = df_min.iloc[0:1830, 9]
df_min.iloc[0:1830, 9] = df_min.iloc[0:1830, 10]
df_min.iloc[0:1830, 10] = df_min.iloc[0:1830, 11]
df_min.iloc[0:1830, 15] = df_min.iloc[0:1830, 16]
df_min.drop([1,4,6,10,11,12,13,14,16], axis=1, inplace=True)
df_min.columns = ["Nominal Date","Wavelength (nm)","Instrument Mode","Spectral Irradiance",
                  "Instrument Uncertainty","Measurement Precision","Measurement Stability","Quality Flag"]
df_min
```

```python
df_min.drop(df_min[df_min['Quality Flag'] == 1].index, inplace=True)
df_min.reset_index(drop=True, inplace=True)
df_min
```

```python
plt.figure(figsize=(25,15))
plt.scatter(df_min['Wavelength (nm)'],df_min['Spectral Irradiance'])
plt.xlabel("Wavelength (nm)", fontsize = 25)
plt.ylabel("Spectral Irradiance (e13.2 W/m²/nm)", fontsize = 25)
plt.title("Solar Spectral Irradiance vs Wavelength", fontsize = 30, fontweight='bold')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.show()
```

## Plank's Function with Original Constants

```python
c1 = 2 * (cs.h) * ((cs.c)**2)
c2 = ((cs.h) * (cs.c)) / ((cs.k) * 5773)
def plank(λ):
    return (c1 * (λ**(-5))) / ((np.e)**(c2/λ) - 1)

B_plank = plank(df_min["Wavelength (nm)"]*(10**(-9)))
```

```python
plt.figure(figsize=(25,15))
plt.plot(df_min['Wavelength (nm)'],B_plank*(10**(-13.2)),color = "red",linewidth = 2.5)
plt.scatter(df_min['Wavelength (nm)'],df_min['Spectral Irradiance'])
plt.xlabel("Wavelength (nm)", fontsize = 25)
plt.ylabel("Spectral Irradiance (e13.2 W/m²/nm)", fontsize = 25)
plt.title("SSI and Plank's Law Comparison", fontsize = 30, fontweight='bold')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.show()
```

## Plank's Function Fit

```python
def plank_fit(λ,a,b):
    return (a * (λ**(-5))) / ((np.e)**(b/λ) - 1)

p0 = [10**(-16), 10**(-6)]

params, cov = fit.curve_fit(plank_fit, df_min['Wavelength (nm)']*(10**(-9)), 
                            df_min['Spectral Irradiance']*(10**13.2), p0)
```

### Test Dataset

```python
wavelength_test = np.linspace(1, 2500, 2500)
B_test = plank_fit(wavelength_test*(10**(-9)), params[0], params[1])

plt.figure(figsize=(25,15))
plt.plot(wavelength_test,B_test*(10**(-13.2)),color = "violet",linewidth = 2.5)
plt.xlabel("Wavelength (nm)", fontsize = 25)
plt.ylabel("Spectral Irradiance (e13.2 W/m²/nm)", fontsize = 25)
plt.title("Plank's Fitted Curve under Test SSI", fontsize = 30, fontweight='bold')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.show()
```

### Real Dataset

```python
B_est = plank_fit(df_min['Wavelength (nm)']*(10**(-9)), params[0], params[1])

plt.figure(figsize=(25,15))
plt.plot(df_min['Wavelength (nm)'],B_est*(10**(-13.2)),color = "chocolate",linewidth = 2.5)
plt.scatter(df_min['Wavelength (nm)'],df_min['Spectral Irradiance'])
plt.xlabel("Wavelength (nm)", fontsize = 25)
plt.ylabel("Spectral Irradiance (e13.2 W/m²/nm)", fontsize = 25)
plt.title("Plank's Fitted Curve under Real SSI", fontsize = 30, fontweight='bold')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.show()
```

## Error Analysis

### Residual Plot

```python
residuals = df_min['Spectral Irradiance']*(10**13.2) - B_est

plt.figure(figsize=(10,5))
plt.scatter(df_min['Wavelength (nm)'],residuals*(10**(-13.2)),s=2,color="green")
plt.xlabel("Wavelength (nm)")
plt.ylabel("Residuals of Spectral Irradiance (e13.2 W/m²/nm)")
plt.title("Residual Plot of Plank's Fitted Curve")
plt.show()
```

### χ² Calculation

```python
std_dev = np.std(df_min['Spectral Irradiance']*(10**13.2))
chi_squared = np.sum((residuals/std_dev) ** 2)
reduced_chi_squared = np.sum((residuals/std_dev) ** 2) / (len(df_min['Wavelength (nm)'] - 2))
print(f"χ² = {chi_squared}")
print(f"reduced χ² = {reduced_chi_squared}")
```
