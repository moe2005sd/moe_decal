## Plank's Function with Original Constants

```python
c1 = 2 * (cs.h) * ((cs.c)**2)
c2 = ((cs.h) * (cs.c)) / ((cs.k) * 5773)
def plank(Î»):
    return (c1 * (Î»**(-5))) / ((np.e)**(c2/Î») - 1)

B_plank = plank(df_max["Wavelength (nm)"]*(10**(-9)))
```

```python
plt.figure(figsize=(25,15))
plt.plot(df_max['Wavelength (nm)'],B_plank*(10**(-13.2)),color = "red",linewidth = 2.5)
plt.scatter(df_max['Wavelength (nm)'],df_max['Spectral Irradiance'])
plt.xlabel("Wavelength (nm)", fontsize = 25)
plt.ylabel("Spectral Irradiance (e13.2 W/mÂ²/nm)", fontsize = 25)
plt.title("SSI and Plank's Law Comparison", fontsize = 30, fontweight='bold')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.show()
```

## Plank's Function Fit

```python
def plank_fit(Î»,a,b):
    return (a * (Î»**(-5))) / ((np.e)**(b/Î») - 1)

p0 = [10**(-16), 10**(-6)]

params, cov = fit.curve_fit(plank_fit, df_max['Wavelength (nm)']*(10**(-9)), 
                            df_max['Spectral Irradiance']*(10**13.2), p0)
```

### Test Dataset

```python
wavelength_test = np.linspace(1, 2500, 2500)
B_test = plank_fit(wavelength_test*(10**(-9)), params[0], params[1])

plt.figure(figsize=(25,15))
plt.plot(wavelength_test,B_test*(10**(-13.2)),color = "violet",linewidth = 2.5)
plt.xlabel("Wavelength (nm)", fontsize = 25)
plt.ylabel("Spectral Irradiance (e13.2 W/mÂ²/nm)", fontsize = 25)
plt.title("Plank's Fitted Curve under Test SSI", fontsize = 30, fontweight='bold')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.show()
```

### Real Dataset

```python
B_est = plank_fit(df_max['Wavelength (nm)']*(10**(-9)), params[0], params[1])

plt.figure(figsize=(25,15))
plt.plot(df_max['Wavelength (nm)'],B_est*(10**(-13.2)),color = "chocolate",linewidth = 2.5)
plt.scatter(df_max['Wavelength (nm)'],df_max['Spectral Irradiance'])
plt.xlabel("Wavelength (nm)", fontsize = 25)
plt.ylabel("Spectral Irradiance (e13.2 W/mÂ²/nm)", fontsize = 25)
plt.title("Plank's Fitted Curve under Real SSI", fontsize = 30, fontweight='bold')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.show()
```

## Error Analysis

### Residual Plot

```python
residuals = df_max['Spectral Irradiance']*(10**13.2) - B_est

plt.figure(figsize=(10,5))
plt.scatter(df_max['Wavelength (nm)'],residuals*(10**(-13.2)),s=2,color="green")
plt.xlabel("Wavelength (nm)")
plt.ylabel("Residuals of Spectral Irradiance (e13.2 W/mÂ²/nm)")
plt.title("Residual Plot of Plank's Fitted Curve")
plt.show()
```

### Ï‡Â² Calculation

```python
std_dev = np.std(df_max['Spectral Irradiance']*(10**13.2))
chi_squared = np.sum((residuals/std_dev) ** 2)
reduced_chi_squared = np.sum((residuals/std_dev) ** 2) / (len(df_max['Wavelength (nm)'] - 2))
print(f"Ï‡Â² = {chi_squared}")
print(f"reduced Ï‡Â² = {reduced_chi_squared}")
```
